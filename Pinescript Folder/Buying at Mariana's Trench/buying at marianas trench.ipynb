```python
//@version=5
indicator("SBMT - Sunstoic's Buying at Marias Trench", overlay=false)

// User Input for Periods
cciPeriod = input.int(14, title="IndOne Period")
rsiPeriod = input.int(5, title="IndTwo Period")
wrPeriod = input.int(14, title="IndThree Period")

// Heikin Ashi Close
haTicker = ticker.heikinashi(syminfo.ticker)
haClose = request.security(haTicker, timeframe.period, close)

// CCI Calculation
IndOne = ta.cci(haClose, cciPeriod)

// RSI Calculation
IndTwo = ta.rsi(haClose, rsiPeriod)

// %R Calculation
IndThree = ta.wpr(wrPeriod)

// Define Threshold Levels
level1 = -150
level2 = -180
level3 = -200
level4 = -250
level5 = -300
level6 = 150
level7 = 180
level8 = 200
level9 = 250
level10 = 300

// IndTwo and IndThree Levels
rsiUpper = 80
rsiLower = 20
wrUpper = -10
wrLower = -90

// Determine Buy Signal Color
color buyColor = color.gray
if IndOne < level1
    buyColor := color.blue
if IndOne < level2
    buyColor := color.green
if IndOne < level3
    buyColor := color.yellow
if IndOne < level4
    buyColor := color.orange
if IndOne < level5
    buyColor := color.red

// Buy Signal Condition
buySignal = IndOne < level1

// Plot Buy Signal with Color Variation
plotshape(buySignal, location=location.bottom, color=buyColor, style=shape.labelup, title="Buy Signal")

// Plot CCI Indicator
plot(IndOne, title="IndOne", color=color.blue)
hline(level1, "Level -150", color=color.blue)
hline(level2, "Level -180", color=color.green)
hline(level3, "Level -200", color=color.yellow)
hline(level4, "Level -250", color=color.orange)
hline(level5, "Level -300", color=color.red)
hline(level6, "Level 150", color=color.blue)
hline(level7, "Level 180", color=color.green)
hline(level8, "Level 200", color=color.yellow)
hline(level9, "Level 250", color=color.orange)
hline(level10, "Level 300", color=color.red)

// Plot IndTwo Indicator
plot(IndTwo, title="IndTwo", color=color.purple)
hline(rsiUpper, "IndTwo level 80", color=color.red)
hline(rsiLower, "IndTwo level 20", color=color.green)

// Plot IndThree Indicator
plot(IndThree, title="IndThree", color=color.orange)
hline(wrUpper, "IndThree level -10", color=color.red)
hline(wrLower, "IndThree level -90", color=color.green)

// Background Highlights for Buy Zones
bgcolor(buySignal ? buyColor : na, transp=90)
```
