## The Law of Large Numbers (LLN)

The **Law of Large Numbers (LLN)** is one of the most important concepts in probability and statistics.  
It explains why averages become stable when an experiment is repeated many times.

---

## 1. What Is the Law of Large Numbers?

The Law of Large Numbers states:

> When a random experiment is repeated many times,  
> the **sample average** will get closer and closer to the **true expected value**.

Examples:
- Roll a die many times → mean approaches $3.5$ (from the mean of 1/6 probability outcome)
- Flip a fair coin many times → proportion of heads approaches $0.5$ (from the mean of 1/2 probability outcome) 
- More data leads to more accurate averages

---

## 2. Why It Matters

LLN explains why:

- Casinos profit in the long run  
- Insurance companies can price risk accurately  
- Traders measure average returns  
- Scientists reduce noise using repeated measurements  
- Machine learning improves with large datasets  

LLN shows that data becomes reliable only when the sample size is large.

---

## 3. Example: Rolling a 6-Sided Die

Expected value:

$$
E[X] = \frac{1 + 2 + 3 + 4 + 5 + 6}{6} = 3.5
$$

Small sample → unstable  
Large sample → stabilizes near $3.5$

## or

The expected value (mean) of a fair 6-sided die can be calculated using:

$$
E[X] = \sum_{i=1}^{6} x_i \cdot P(X = x_i)
$$

Since the die is fair, each outcome has probability $P(X = x_i) = \frac{1}{6}$:

$$
E[X] = 1 \cdot \frac{1}{6} + 2 \cdot \frac{1}{6} + 3 \cdot \frac{1}{6} + 4 \cdot \frac{1}{6} + 5 \cdot \frac{1}{6} + 6 \cdot \frac{1}{6} = \frac{21}{6} = 3.5
$$

According to the Law of Large Numbers, if we roll the die many times, the sample mean will approach $3.5$.


---

## 4. Example: Flipping a Fair Coin

Represent:
- Heads = 1  
- Tails = 0  

Expected value:

$$
E[X] = 0.5
$$

With many flips, the proportion of heads approaches $0.5$.

---

## 5. Sample Mean vs. True Mean

### Sample Mean
$$
\bar{X}_n = \frac{X_1 + X_2 + \cdots + X_n}{n}
$$

### True Mean
$$
\mu = E[X]
$$

LLN states:

$$
\bar{X}_n \rightarrow \mu \quad \text{as } n \to \infty
$$

---

## 6. Two Types of the Law of Large Numbers

### A. Weak Law (WLLN)  
Convergence *in probability*:

$$
\bar{X}_n \xrightarrow{P} \mu
$$

The more samples you have, the less likely the sample mean is far from the true mean.

---

### B. Strong Law (SLLN)  
Convergence *almost surely*:

$$
\bar{X}_n \xrightarrow{a.s.} \mu
$$

With probability 1, the sample mean will approach the true mean.

---

## 7. What the LLN Does *Not* Say

Common misconceptions:

- It does not mean the next flip will “balance out” previous ones.  
- It does not mean small samples match the expected value.  
- It does not work in the short term.  
- It does not guarantee profit.  

LLN only works when sample size becomes large.

---

## 8. Why LLN Works: Intuition

Random values include:
- High values  
- Low values  

As repetitions increase, highs and lows cancel out.  
This forces the average to approach the true mean.

Small sample = unstable  
Large sample = stable  

---

## 9. LLN in Real Life

### Casinos
- House edge is small  
- Millions of bets guarantee profit

### Insurance
- Predicting claims and risk  

### Trading
- Average return stabilizes over many trades  

### Machine Learning
- More data → more reliable learning  

### Science
- Repeated experiments reduce measurement noise

---

## 10. Mathematical Formulation

Let $X_1, X_2, ..., X_n$ be i.i.d. random variables with expected value $\mu$.

$$
\lim_{n \to \infty} \bar{X}_n = \mu
$$

Weak LLN → convergence in probability  
Strong LLN → almost sure convergence  

---

## 11. Requirements for LLN to Hold

LLN requires:

- Independent, identically distributed samples  
- Finite expected value  
- Large sample size  

LLN may fail for heavy-tailed distributions.

---

## 12. Visual Understanding

An LLN plot typically shows:

- Large fluctuations when sample size is small  
- Smoother convergence as sample size increases  
- Mean approaching a horizontal line (true expected value)

---

## 13. Summary Table

| Concept | Meaning |
|--------|---------|
| LLN | Average approaches true expected value |
| Weak LLN | Convergence in probability |
| Strong LLN | Almost sure convergence |
| Applies to | Coins, dice, trading, science |
| Requires | i.i.d samples, finite mean |
| LLN does not say | Short-term correction, guaranteed profit |

---

## 14. Final Takeaway

The Law of Large Numbers explains why averages become accurate and reliable  
when the number of trials becomes large.

It is a foundation of probability, statistics, finance, science, and machine learning.


## Simulation of LLN

```python
# ---------------------------------------------------------
# Law of Large Numbers Demonstration
# Die (6 sides) and Coin (2 sides)
# ---------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# 1. SIX-SIDED DIE SIMULATION
# ---------------------------------------------------------

n_trials = 1000  # number of rolls/flips

# Roll a 6-sided die (values 1–6)
die_rolls = np.random.randint(1, 7, size=n_trials)

# Compute cumulative means
die_mean = np.cumsum(die_rolls) / np.arange(1, n_trials + 1)

# Expected value of a fair die
die_expected = (1 + 2 + 3 + 4 + 5 + 6) / 6  # = 3.5

# Plot
plt.figure(figsize=(12, 5))
plt.plot(die_mean, label="Sample Mean (Die)")
plt.axhline(die_expected, color="red", linestyle="--", label="Expected Value = 3.5")
plt.title("Law of Large Numbers — 6-Sided Die")
plt.xlabel("Number of Rolls")
plt.ylabel("Mean Result")
plt.grid(True)
plt.legend()
plt.show()


# ---------------------------------------------------------
# 2. TWO-SIDED COIN SIMULATION
# ---------------------------------------------------------

# Represent Heads = 1, Tails = 0
coin_flips = np.random.randint(0, 2, size=n_trials)

# Cumulative mean (approaches 0.5)
coin_mean = np.cumsum(coin_flips) / np.arange(1, n_trials + 1)

coin_expected = 0.5  # Expected mean of fair coin

# Plot
plt.figure(figsize=(12, 5))
plt.plot(coin_mean, label="Sample Mean (Coin)")
plt.axhline(coin_expected, color="red", linestyle="--", label="Expected Value = 0.5")
plt.title("Law of Large Numbers — Fair Coin")
plt.xlabel("Number of Flips")
plt.ylabel("Mean (Probability of Heads)")
plt.grid(True)
plt.legend()
plt.show()


# ---------------------------------------------------------
# FINAL PRINT RESULTS
# ---------------------------------------------------------

print("FINAL RESULTS")
print("--------------------------------------")
print(f"Die final sample mean: {die_mean[-1]:.4f} (Expected: 3.5)")
print(f"Coin final sample mean: {coin_mean[-1]:.4f} (Expected: 0.5)")
```
