This is a candle patterns:

If the current candle (position 0) low is lower than position -1’s low AND the close of the current candle (position 0) is higher than the open of position -1 candle. Then mark this as a bullish pattern.

If the current candle (position 0) high is higher than position -1’s high AND the close of the current candle (position 0) is lower than the open of position -1 candle. Then mark this as a bearish pattern.
