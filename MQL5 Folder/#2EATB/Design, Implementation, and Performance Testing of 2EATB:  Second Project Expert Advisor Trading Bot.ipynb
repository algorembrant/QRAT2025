##  Discussion
**Why Exness**
Exness is preferred for algorithmic trading because it fully supports Expert Advisors (EAs) on MT4 and MT5, allowing automated strategies on live accounts with withdrawals handled like regular trades. Its Cent accounts let traders test strategies with minimal risk, and third-party EAs are explicitly permitted, offering flexibility for both beginners and experienced traders (Exness Service, n.d.).

**Why Python**

Python is widely used in algorithmic trading for backtesting because it is easy to learn, has powerful libraries (like pandas, NumPy, TA-Lib, backtrader), allows fast data manipulation and analysis, and supports rapid prototyping of trading strategies with clean, readable code. It also integrates well with APIs for historical data and broker execution. As for this study Python 3.13 version is used instead of other upgraded version due to a reason that MT5 only suports Python 3.13 and below as of the moment when this was conducted.

**Why MT5**

MetaTrader 5 (MT5) is widely used for backtesting because it provides reliable historical price data and integrates seamlessly with Python via the MT5-Python API, allowing traders to fetch data, analyze it with libraries like pandas and NumPy, and test automated strategies efficiently. Python adds flexibility for complex analysis, optimization, and machine learning, while MT5 handles execution and order placement. This combination enables robust backtesting and smooth transition from testing to live trading (Documentation on MQL5: Python Integration, n.d.)

**Spread**

The spread is the difference between the ask price (buy price) and the bid price (sell price) that a broker quotes for a currency pair or asset. It is essentially an instant cost built into the market price. Spreads can be fixed (always the same) or variable (changing with market volatility). You pay the spread immediately when entering a trade because it is part of the market execution and does not require a separate fee.

**Commission**

A commission is a separate fee charged by the broker for executing a trade. It is usually calculated as a percentage of trade size, a fixed amount per lot, or sometimes as a percentage of profit or loss in backtests. Unlike spread, commission does not affect the entry price; instead, it is added to or subtracted from your account balance independently. Some brokers offer zero-spread accounts but compensate by charging commissions.
