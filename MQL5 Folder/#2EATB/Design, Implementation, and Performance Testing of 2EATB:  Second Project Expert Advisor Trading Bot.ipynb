##   Introoduction

The spread is the difference between the ask price (buy price) and the bid price (sell price) that a broker quotes for a currency pair or asset. It’s essentially an instant “cost” built into the market price. Spreads can be fixed (same always) or variable (changes with market volatility). You pay the spread immediately when entering a trade; it is part of the market execution and does not require a separate fee.

## Introduction

The spread is the difference between the ask price (buy price) and the bid price (sell price) that a broker quotes for a currency pair or asset. It’s essentially an instant “cost” built into the market price. Spreads can be fixed (same always) or variable (changes with market volatility). You pay the spread immediately when entering a trade; it is part of the market execution and does not require a separate fee.
