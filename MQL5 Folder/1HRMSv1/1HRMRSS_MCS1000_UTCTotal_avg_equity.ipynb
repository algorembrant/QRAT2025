```python
import pandas as pd
import plotly.graph_objects as go

# ----------------------------------------
# 1. LOAD DATA
# ----------------------------------------
file_path = "1HRMRSS_MCS1000 - total.csv"
df = pd.read_csv(file_path)

# ----------------------------------------
# 2. REMOVE UNWANTED COLUMNS
# ----------------------------------------
if "Equity Start" in df.columns:
    df = df.drop(columns=["Equity Start"])
df = df.loc[:, ~df.columns.str.contains("^Unnamed")]

# ----------------------------------------
# 3. DETECT PLOT COLUMNS
# ----------------------------------------
value_columns = df.select_dtypes(include="number").columns.tolist()
if "index" in value_columns:
    value_columns.remove("index")
x_axis = df["index"] if "index" in df.columns else df.index

# ----------------------------------------
# 4. PLOTLY INTERACTIVE GRAPH
# ----------------------------------------
fig = go.Figure()

# Add equity curves
for col in value_columns:
    fig.add_trace(
        go.Scatter(
            x=x_axis,
            y=df[col],
            mode="lines",
            name=col,
            line=dict(width=2)
        )
    )

# Add starting equity line at 1000
fig.add_trace(
    go.Scatter(
        x=[x_axis.min(), min(500, x_axis.max())],  # from 0 to 500 (or max available)
        y=[1000, 1000],
        mode="lines",
        name="Starting Equity",
        line=dict(color="black", width=2, dash="dash")
    )
)

# Update layout
fig.update_layout(
    title="Average Equity Curves from 1000 MCS runs each 6000-1H-candle-based Datafeed",
    xaxis_title="Trades",
    yaxis_title="Equity",
    template="plotly_white",
    hovermode="x unified",
    height=800,
    autosize=True,
    xaxis=dict(range=[0, 500]),
    yaxis=dict(range=[0, 2000]),
    margin=dict(l=60, r=40, t=80, b=60)
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
from io import StringIO

# ---------------------------
# 1. CSV Data
# ---------------------------
csv_data = """
Rank,Session Start,Session End,Equity,Trades
1,20:00,21:00,812.3243254,130
2,19:00,20:00,857.5294073,170
3,14:00,15:00,603.1843095,202
4,18:00,19:00,798.1902598,221
5,15:00,16:00,650.1141337,228
6,13:00,14:00,1175.56929,230
7,17:00,18:00,1010.071902,243
8,16:00,17:00,708.1158584,245
9,12:00,13:00,1495.813783,309
10,1:00,2:00,524.7783349,355
11,11:00,12:00,1313.135488,361
12,7:00,8:00,512.7152816,384
13,8:00,9:00,380.9657206,386
14,0:00,1:00,1049.417423,386
15,6:00,7:00,577.5135022,388
16,2:00,3:00,443.6543421,393
17,5:00,6:00,515.5098371,397
18,10:00,11:00,1216.246919,398
19,9:00,10:00,437.7753629,399
20,3:00,4:00,195.9143007,427
21,4:00,5:00,736.8399835,428
"""

# ---------------------------
# 2. Load DataFrame
# ---------------------------
df = pd.read_csv(StringIO(csv_data))

# Convert 'Session Start' to datetime for sorting
df['Session Start Time'] = pd.to_datetime(df['Session Start'], format='%H:%M')
df = df.sort_values('Session Start Time').reset_index(drop=True)
df['Session Start'] = df['Session Start Time'].dt.strftime('%H:%M')

# ---------------------------
# 3. Compute Performance Score
# ---------------------------
w_E, w_T = 0.5, 0.5
df['Equity_norm'] = (df['Equity'] - df['Equity'].min()) / (df['Equity'].max() - df['Equity'].min())
df['Trades_norm'] = (df['Trades'] - df['Trades'].min()) / (df['Trades'].max() - df['Trades'].min())
df['Score'] = w_E * df['Equity_norm'] + w_T * df['Trades_norm']

# Rank based on Score if needed
df['ScoreRank'] = df['Score'].rank(method='dense', ascending=False).astype(int)

# ---------------------------
# 4. 3D Bubble Chart
# ---------------------------
fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x=df['Session Start'],          
    y=df['Equity'],                 
    z=df['Trades'],                 
    mode='markers+text',
    marker=dict(
        size=df['Score'] * 40,        # Bubble size proportional to Score
        color=df['Score'],            # Color based on Score
        colorscale='Magma',        # Color scheme
        showscale=True,
        reversescale=True,
        colorbar=dict(title='Performance Score')
    ),
    text=df['ScoreRank'],            # Text inside circle = ScoreRank
    textposition='top center',
    hovertemplate=(
        'Session: %{x}<br>'
        'Equity: %{y}<br>'
        'Trades: %{z}<br>'
        'Performance Score: %{customdata[0]:.3f}<br>'
        'Score Rank: %{text}'
    ),
    customdata=df[['Score']]
))

# ---------------------------
# 5. Layout
# ---------------------------
fig.update_layout(
    title='Session Performance: Equity and Trades',
    scene=dict(
        xaxis_title='Session',
        yaxis_title='Equity',
        zaxis_title='Number of Trades'
    ),
    width=800,
    height=650,
    margin=dict(l=20, r=20, t=60, b=20),
    template='plotly_white'
)

# ---------------------------
# 6. Show Plot
# ---------------------------
fig.show()
```
